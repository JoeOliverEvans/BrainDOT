## Looking at the imaging volume

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import mat73
import scipy.io as sio
```

```python
images = sio.loadmat('test_processed.mat')
recon2 = images['recon2']
```

```python
recon2.shape
```

```python
np.min(recon2)
```

```python
colors = np.empty(recon2.shape, dtype=object)
colors[:] = 'black'
colors[recon2>0.001] = 'red'
colors[recon2<-0.002] = 'blue'
```

```python
ax = plt.figure().add_subplot(projection='3d')
ax.voxels(recon2[:,:,:,0], facecolors=colors[:,:,:,0], alpha=1)

plt.show()
```
