```python
import numpy as np
import mat73
import scipy.io as sio
from DeepUNET import UNet
```

```python
dataset_names = ['CCW1', 'CCW2', 'CW1']
model_string = 'BiggerDatasetDeeper'
```

```python
data = {}
for name in dataset_names:
    dataset = sio.loadmat(f'{model_string}/{name}/{name}_processed.mat')
    for key in dataset.keys():
        if key not in data.keys():
            data[key] = dataset[key]
        else:
            data[key] = [*data[key], *dataset[key]]
```
