## Looking at the imaging volume

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import mat73
```

```python
badata_HbOvol = mat73.loadmat('TestingOutputs/badata_HbOvol_file.mat')
badata_HbOvol = badata_HbOvol['badata_HbOvol']
```

```python
badata_HbOvol.shape
```

```python
np.min(badata_HbOvol)
```

```python
colors = np.empty(badata_HbOvol.shape, dtype=object)
colors[:] = 'black'
colors[badata_HbOvol>0.001] = 'red'
colors[badata_HbOvol<-0.002] = 'blue'
```

```python
ax = plt.figure().add_subplot(projection='3d')
ax.voxels(badata_HbOvol[:,:,:,15], facecolors=colors[:,:,:,15], alpha=1)

plt.show()
```
