## Looking at the imaging volume

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import mat73
import scipy.io as sio
```

```python
img = sio.loadmat('Datasets/FirstModel/test_processed.mat')
noisy_img = img['noisy_images']
recon_img = img['recon2']
clean_img = img['eval_images']
```

```python
i=3
```

```python
ax = plt.figure().add_subplot(projection='3d')
ax.voxels(clean_img[:,:,:,i], alpha=1)

plt.show()
```

```python
ax = plt.figure().add_subplot(projection='3d')
ax.voxels(abs(noisy_img[:,:,:,i])>8, alpha=1)

plt.show()
```

```python
ax = plt.figure().add_subplot(projection='3d')
ax.voxels(abs(recon_img[:,:,:,i])>4, alpha=1)

plt.show()
```
